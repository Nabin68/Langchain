```python
from langchain_core.tools import tool
from langchain_cohere import ChatCohere
from dotenv import load_dotenv
from langchain_core.messages import HumanMessage
import requests
```

Step1:Tool Creation

```python
#First lets create a tool

@tool
def Multiplication_tool(a:int,b:int)->int:
    """"Giiven 2 numbers a and b ,this tool return their product"""
    return a*b
```

```python
Multiplication_tool.invoke({"a":12,"b":4})
```

```python
load_dotenv()
llm=ChatCohere(model="command-a-03-2025")
```

Step2:Tool Binding

```python
llm_with_tools=llm.bind_tools([Multiplication_tool])
```

Step3:Tool Calling

```python
query=HumanMessage("Whats 5 multiplied with 4?")
messages=[query]
```

```python
result=llm_with_tools.invoke(messages)
messages.append(result)
```

```python
messages
```

```python
tool_result=Multiplication_tool.invoke(result.tool_calls[0])
messages.append(tool_result)
```
