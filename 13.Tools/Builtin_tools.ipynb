Build in DuckDuckGo Search

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool=DuckDuckGoSearchRun()

result=search_tool.invoke("IPL news")

print(result)
```

Builtin Shell Tool

```python
from langchain_community.tools import ShellTool

shell_tool=ShellTool()

results=shell_tool.invoke("ls") #whoami -> name of the the system or user, in this way we can run diff shell commands

print(results)
```
