```python
from langchain.tools import tool

#Custom tools

@tool
def add(a:int,b:int)->int:
    """Add two numbers"""
    return a+b

@tool
def subtract(a:int,b:int)->int:
    """Subtract two numbers"""
    return a-b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add,subtract]
```

```python
toolkit=MathToolkit()
tools=toolkit.get_tools()

for tool in tools:
    print(tool.name,"=>",tool.description)
```
